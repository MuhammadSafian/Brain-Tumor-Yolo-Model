```python
# load library
from ultralytics import YOLO
```

```python
# select model
model =YOLO("yolov8n.pt")
```

```python
# yaml file
yaml_path ="/content/medical-image-dataset-brain-tumor-detection/BrainTumor/BrainTumorYolov8/data.yaml"
```

```python
#Train model
model.train(
    data =yaml_path,
    epochs =100,
    imgsz =640,
    batch =16,
    patience =20,
    pretrained =True
)
```

```python
# Path to test image
test_image = "/content/medical-image-dataset-brain-tumor-detection/BrainTumor/BrainTumorYolov8/test/images/2946_jpg.rf.855119c7ea912804400e339a720afcc0.jpg"
results = model.predict(source=test_image)

# Show result
results[0].plot()
```
